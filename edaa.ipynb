```python
time=pd.DataFrame()
time['Time_in_string']=eda['Time']
```

```python
time['Time_in_datetime'] = pd.to_datetime(time['Time_in_string'], format='%H:%M:%S')

time['Hour'] = time['Time_in_datetime'].dt.hour
time['Minute'] = time['Time_in_datetime'].dt.minute
time['Second'] = time['Time_in_datetime'].dt.second
```

```python
time.head(4)
```

```python
time['Hour']=time['Hour'].replace([])
```

```python
time.loc[(time['Hour'] >= 0) & (time['Hour'] <= 3), 'Hour'] = 1
time.loc[(time['Hour'] > 3) & (time['Hour'] < 6), 'Hour'] = 2
time.loc[(time['Hour'] >= 6) & (time['Hour'] < 12), 'Hour'] = 3
time.loc[(time['Hour'] >= 12) & (time['Hour'] < 16), 'Hour'] = 4
time.loc[(time['Hour'] >= 16) & (time['Hour'] < 18), 'Hour'] = 5
time.loc[(time['Hour'] >= 18) & (time['Hour'] < 24), 'Hour'] = 6
```
